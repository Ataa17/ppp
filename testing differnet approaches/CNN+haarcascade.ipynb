```python
!pip install opencv-python
!pip install scikit-learn
!pip install matplotlib
!pip install tensorflow

import numpy as np
import matplotlib.pyplot as plt
import cv2
from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization
from tensorflow.keras.optimizers import Adam
import urllib.request
import os

print("All libraries installed and imported!")

# Download Haar cascade for face detection
haar_cascade_url = 'https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml'
haar_cascade_path = 'haarcascade_frontalface_default.xml'

if not os.path.exists(haar_cascade_path):
    print("Downloading Haar cascade...")
    urllib.request.urlretrieve(haar_cascade_url, haar_cascade_path)
    print("Haar cascade downloaded!")

# Load the Haar cascade
face_cascade = cv2.CascadeClassifier(haar_cascade_path)

# Load LFW dataset
# Add data augmentation to improve generalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator

print("Applying data augmentation...")
datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    horizontal_flip=True,
    zoom_range=0.2,
    shear_range=0.2,
    fill_mode='nearest'
)

# Alternative training with data augmentation (uncomment to use)
# history = model.fit(
#     datagen.flow(X_train, y_train, batch_size=32),
#     steps_per_epoch=len(X_train) // 32,
#     epochs=50,
#     validation_data=(X_test, y_test),
#     callbacks=[early_stopping, reduce_lr],
#     verbose=1
# )
lfw_people = fetch_lfw_people(min_faces_per_person=20, resize=0.4)
X = lfw_people.images  # image data
y = lfw_people.target  # target labels
target_names = lfw_people.target_names  # label names

print("Shape of X:", X.shape)
print("Number of classes:", len(target_names))
print("People in dataset:", target_names)

# Normalize and prepare data
X = X / 255.0  # Normalize pixel values
X = X.reshape(X.shape[0], X.shape[1], X.shape[2], 1)  # Add channel dimension
y_cat = to_categorical(y, len(target_names))  # One-hot encode

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

# Build a simpler, more appropriate model for this dataset
print("Building simplified CNN model...")
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau

model = Sequential()

# Simple architecture more suitable for small images (50x37)
model.add(Conv2D(16, (3, 3), activation='relu', padding='same', input_shape=X_train.shape[1:]))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(32, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), activation='relu', padding='same'))
model.add(BatchNormalization())
model.add(MaxPooling2D((2, 2)))
model.add(Dropout(0.25))

# Global Average Pooling instead of Flatten to reduce parameters
from tensorflow.keras.layers import GlobalAveragePooling2D
model.add(GlobalAveragePooling2D())

# Smaller dense layers
model.add(Dense(64, activation='relu', kernel_regularizer=l2(0.01)))
model.add(Dropout(0.5))
model.add(Dense(len(target_names), activation='softmax'))

# Compile with appropriate settings
model.compile(
    optimizer=Adam(learning_rate=0.001),  # Start with higher learning rate
    loss='categorical_crossentropy', 
    metrics=['accuracy']
)

print("Model Summary:")
model.summary()

# Define callbacks
early_stopping = EarlyStopping(
    monitor='val_accuracy', 
    patience=10, 
    restore_best_weights=True,
    verbose=1
)
reduce_lr = ReduceLROnPlateau(
    monitor='val_loss', 
    factor=0.5, 
    patience=5, 
    min_lr=1e-6,
    verbose=1
)

# Train the model with proper data
print("Training model...")
history = model.fit(
    X_train, y_train, 
    epochs=50,
    batch_size=16,  # Smaller batch size for better convergence
    validation_data=(X_test, y_test), 
    callbacks=[early_stopping, reduce_lr],
    verbose=1
)

# Evaluate model
loss, accuracy = model.evaluate(X_test, y_test, verbose=0)
print(f"Test accuracy: {accuracy * 100:.2f}%")

# Additional evaluation metrics
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

# Get predictions
y_pred = model.predict(X_test, verbose=0)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true_classes = np.argmax(y_test, axis=1)

print("\nClassification Report:")
print(classification_report(y_true_classes, y_pred_classes, target_names=target_names))

# Plot confusion matrix for first few classes
n_classes_to_show = min(5, len(target_names))
cm = confusion_matrix(y_true_classes, y_pred_classes)
plt.figure(figsize=(10, 8))
sns.heatmap(cm[:n_classes_to_show, :n_classes_to_show], 
            annot=True, fmt='d', cmap='Blues',
            xticklabels=target_names[:n_classes_to_show],
            yticklabels=target_names[:n_classes_to_show])
plt.title('Confusion Matrix (First 5 Classes)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.show()

# Plot training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.title("Training Accuracy")

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.title("Training Loss")
plt.tight_layout()
plt.show()

# Function to preprocess detected face for prediction
def preprocess_face(face_img, target_size=(50, 37)):
    """Preprocess detected face to match LFW dataset format"""
    # Convert to grayscale if needed
    if len(face_img.shape) == 3:
        face_img = cv2.cvtColor(face_img, cv2.COLOR_BGR2GRAY)
    
    # Resize to match LFW dataset dimensions
    face_resized = cv2.resize(face_img, target_size)
    
    # Normalize
    face_normalized = face_resized / 255.0
    
    # Add batch and channel dimensions
    face_processed = face_normalized.reshape(1, target_size[1], target_size[0], 1)
    
    return face_processed

# Function to detect and identify faces
def detect_and_identify_faces(image_path, confidence_threshold=0.5):
    """Detect faces using Haar cascade and identify them using trained model"""
    # Read image
    img = cv2.imread(image_path)
    if img is None:
        print(f"Could not load image: {image_path}")
        return None
    
    # Convert to grayscale for face detection
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # Detect faces
    faces = face_cascade.detectMultiScale(
        gray,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30, 30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
    
    print(f"Found {len(faces)} face(s)")
    
    # Process each detected face
    results = []
    for i, (x, y, w, h) in enumerate(faces):
        # Extract face region
        face_roi = gray[y:y+h, x:x+w]
        
        # Preprocess face for prediction
        face_processed = preprocess_face(face_roi)
        
        # Make prediction
        prediction = model.predict(face_processed, verbose=0)
        confidence = np.max(prediction)
        predicted_class = np.argmax(prediction)
        
        # Get predicted name
        predicted_name = target_names[predicted_class]
        
        # Store results
        result = {
            'face_id': i + 1,
            'bbox': (x, y, w, h),
            'predicted_name': predicted_name,
            'confidence': confidence,
            'is_confident': confidence > confidence_threshold
        }
        results.append(result)
        
        # Draw bounding box and label on image
        color = (0, 255, 0) if confidence > confidence_threshold else (0, 0, 255)
        cv2.rectangle(img, (x, y), (x+w, y+h), color, 2)
        
        # Prepare label text
        if confidence > confidence_threshold:
            label = f"{predicted_name} ({confidence:.2f})"
        else:
            label = f"Unknown ({confidence:.2f})"
        
        # Add text background
        (text_width, text_height), baseline = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.6, 2)
        cv2.rectangle(img, (x, y - text_height - 10), (x + text_width, y), color, -1)
        cv2.putText(img, label, (x, y - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)
    
    return img, results

# Function to test with sample images from dataset
def test_with_sample_images(num_samples=5):
    """Test face detection and recognition with sample images from the dataset"""
    print(f"\nTesting with {num_samples} sample images from the dataset...")
    
    # Select random samples from test set
    sample_indices = np.random.choice(len(X_test), num_samples, replace=False)
    
    plt.figure(figsize=(15, 3 * num_samples))
    
    for i, idx in enumerate(sample_indices):
        # Get sample image and true label
        sample_img = X_test[idx].reshape(X_test.shape[1], X_test.shape[2])
        true_label_idx = np.argmax(y_test[idx])
        true_name = target_names[true_label_idx]
        
        # Convert to uint8 format for display
        sample_img_uint8 = (sample_img * 255).astype(np.uint8)
        
        # Make prediction directly (without Haar cascade since these are cropped faces)
        prediction = model.predict(X_test[idx:idx+1], verbose=0)
        confidence = np.max(prediction)
        predicted_class = np.argmax(prediction)
        predicted_name = target_names[predicted_class]
        
        # Display results
        plt.subplot(num_samples, 2, 2*i + 1)
        plt.imshow(sample_img, cmap='gray')
        plt.title(f"True: {true_name}")
        plt.axis('off')
        
        plt.subplot(num_samples, 2, 2*i + 2)
        plt.imshow(sample_img, cmap='gray')
        color = 'green' if predicted_name == true_name else 'red'
        plt.title(f"Predicted: {predicted_name} ({confidence:.3f})", color=color)
        plt.axis('off')
        
        print(f"Sample {i+1}: True={true_name}, Predicted={predicted_name}, Confidence={confidence:.3f}")
    
    plt.tight_layout()
    plt.show()

# Test with sample images from the dataset
test_with_sample_images(5)

# Function to test with webcam (if available)

# Print instructions for testing with custom images
print("\n" + "="*60)
print("FACE DETECTION AND RECOGNITION SYSTEM READY!")
print("="*60)

print("\nTo display results:")
print("plt.figure(figsize=(12, 8))")
print("plt.imshow(cv2.cvtColor(img_result, cv2.COLOR_BGR2RGB))")
print("plt.axis('off')")
print("plt.title('Face Detection and Recognition Results')")
print("plt.show()")
print("\nDetection results will include:")
print("- Face ID, bounding box coordinates")
print("- Predicted name and confidence score")
print("- Whether the prediction is confident (>50%)")

```
